```python
#Data cleaning - Test for many to many (m..m) relationships
#Is purchase ID column uniuque over all records?
dist_purchase_cnt = len(purchase_data['Purchase ID'].unique())
dist_purchase_cnt == len(purchase_data)
```

```python
#Do any purchases contain more than one item?
    #Look at max item ID, if no commas, then no purchase has >1 item
purchase_data['Item ID'].max()
```

```python
#Do any items have more than one price? 
    #Concatenate item ID, Price into Test column
cols_to_keep = ['Item ID','Price']
dfprices = purchase_data[cols_to_keep].copy()
dfprices['Test'] =dfprices['Item ID'].map(str) + '-' + dfprices['Price'].map(str) 

print("count of distinct Item & Price:" + str(len(dfprices['Test'].unique())) + '\n'
+ "count of Item:" + str(len(dfprices['Item ID'].unique())))
    #If unique count of Item ID & Price == unique count of item ID, then no item has >1 price
    #(although some different items may bear the same price)
```

```python
#Unfortunately, Item ID - Price column contains greater count of unique values than Item ID alone. 
#This implies that at least one item has multiple prices. 
    #Investigating by taking array of unique item-price values,
    #split into new [Item ID] and [Price] fields, should be 183 elements. Group by Item ID and aggregate by Count. Look
    #at Item ID and Price for any item ID with a count > 1.

arrItemPrice = dfprices['Test'].unique()
dfItemPrice = pd.DataFrame(arrItemPrice, columns=['Item-Price'])

#Split Item-Price column back into Item ID and Price columns 
#source: https://stackoverflow.com/questions/14745022/how-to-split-a-dataframe-string-column-into-two-columns
    #retrieved 2021-01-16
    
dfItemPrice[['Item ID', 'Price']] = dfItemPrice['Item-Price'].str.split('-', 1, expand = True)
dfItemPrice.head()
```

```python
#Return count of Item-Price combinations by Item ID
dfItemCnts = dfItemPrice.groupby(['Item ID']).count()

#Filter dfItemCnts by count>1
    #cf, https://www.geeksforgeeks.org/ways-to-filter-pandas-dataframe-by-column-values/
        #retrieved 2021-01-16
df_Result = dfItemCnts[dfItemCnts['Price']>1]
df_Result
```

```python
#Show the distinct prices of items with >1 price
suspectItems = [str(1), str(132), str(30), str(92)]
df_search = dfItemPrice[dfItemPrice['Item ID'].isin(suspectItems)]
df_search
```
